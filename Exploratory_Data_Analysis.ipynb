#  자료정리와 요약
### Tips 데이터셋을 활용한 요일별 손님들의 팁 분포 확인 및 이상치 (유난히 팁 을 많이 주거나 적게 준 손님) 확인을 위한 상자그림(box plot) 그리기
Tips 데이터셋: Seaborn 라이브러리 내장 (1990년대 초반, 미국의 한 레스토랑 웨이터가
약 두 달 반 동안 자신이 받은 팁을 기록한 실제 데이터)

 데이터 분석 목표
- 어느 요일의 박스가 가장 긴가?
- 이상치가 있는가? 어느 요일에 이상치가 많이 있는가?
- 내가 알바생이라면 어느 요일에 일할지? 총수익? 효율성?

```python
import seaborn as sns
import matplotlib.pyplot as plt

#데이터 로드
tips = sns.load_dataset("tips")
print(tips.head())
daily_stats= tips.groupby('day')['total_bill'].agg(['mean'])
print(daily_stats)
sns.boxplot(x='day', y='tip', data=tips)
plt.savefig('ex1_1.png')

#팁의 총합과 평균적인 팁 계산하기
tip_analysis = tips.groupby('day')['tip'].agg(['sum','mean','count','max'])
print(tip_analysis)
```

1. boxplot의 결과를 보면 일요일의 박스가 제일 길다는것을 알 수 있습니다. 이를 통해 일요일의 팁
금액 분포가 제일 길다는 사실을 알 수 있었습니다.
2. 이상치가 제일 많은 요일은 토요일이였으며 특히
높은 금액의 이상치 값이 많이 발견되는것으로 보아 일부 고객에게 높은 금액의 팁을 받았음을 알 수
있습니다.
3. 목요일도 이상치가 발견되긴 하지만 토요일보다는 적게 발견됨을 알 수 있습니다. 따라서
#최종 정리

**일요일**이 더 효율적이라고 할 수 있습니다. 따라서 제가 알바생이라면 노동에 따른 체력소모도
