# 확률과 베이즈 정리
### 타이타닉(Titanic) 생존자 데이터셋을 활용하여 “여성 탑승객은 생존 확률이 높았을까?”를 조건부 확률로 계산하고, 베이즈 정리를 통해 검증
Titanic 데이터셋: Seaborn 라이브러리 내장 (1912년 타이타닉호 침몰 사고 당시 승객들
의 생존 여부, 성별, 나이, 객실 등급 등 다양한 정보를 담고 있는 데이터)

• 데이터 분석 포인트

– 직접 계산한 조건부 확률과 베이즈 정리 결과가 일치함을 확인



```python
import seaborn as sns

#데이터 업로드, 총 탑승객수 구하기
titanic = sns.load_dataset('titanic')
total_passenger=len(titanic)

#event a (생존한 사건)
event_survived = titanic[titanic['survived']==1]
p_survived = len(event_survived)/total_passenger
print(p_survived*100)

#event b (여자인 사건)
event_female = titanic[titanic['sex']== 'female']
p_female = len(event_female)/total_passenger
print(p_female * 100)

#event A X B
event_female_survived= titanic[(titanic['sex']=='female') &
(titanic['survived']==1)]
p_female_survived = len(event_female_survived)/total_passenger
print(p_female_survived*100)

#conditional probability ( 조건부 확률 구하기)

#p(survived|female)= p(surived X female)/ p(female)
p_survived_given_female=p_female_survived/p_female
print(f'전체 생존 확률 : {p_survived*100:.2f}%')
print(f'여성일때 생존 확률 : {p_survived_given_female*100:.2f}%')

#Bayes Theorem (베이즈 정리)

#P(survived|female)=P(female|survived)*P(survived)/P(female)
p_female_given_survived=len(event_female_survived)/len(event_survived
)
bayes_result=p_female_given_survived*p_survived/p_female
print(f'베이즈 정리로 계산한 확률 : {bayes_result*100:.2f}%')
```

위의 타이타닉 데이터 분석을 해본 결과

그중 **여성의 생존확률은 74프로**정도로 **여성인 경우 생존확률이 높았**음을 알 수 있다.

**남성이 여성보다는 사망률이 높을 수 있다**는 점과 어린이와 노약자 여성을 먼저 대피 시키는 대피
